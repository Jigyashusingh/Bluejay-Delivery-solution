```python
# First we will import the pandas library to read and analyze the excel file
import pandas as pd
```

```python
# now we will store the path of the file in the variable named file_path.
file_path = "/content/Assignment_Timecard.xlsx"

# then we are reading the excel file and making a data frame out of it.
df = pd.read_excel(file_path)
```

```python
# So in first task we are printing the name and position of employees who have worked for 7 consecutive days.
# creating  a function that take the data frame and giving us the names of the employees for our above task.

def find_employees_with_consecutive_days(df):
    # Converting the 'Time' column to datetime . This conversion is necessary because it allows us to perform datetime operations on the data.
    df['Time'] = pd.to_datetime(df['Time'], errors='coerce')

    # Group by 'Employee Name' and filter employees with 7 consecutive days
    for employee, group in df.groupby('Employee Name'):

        consecutive_days = 0      # Initialize a counter for consecutive days
        prev_date = None    # Initialize a variable to store the previous date

        for idx, row in group.iterrows():
            if prev_date is not None:
                # Check if the difference between the current date and the previous date is exactly 1 day
                if (row['Time'] - prev_date).days == 1:
                    consecutive_days += 1   # Increment the consecutive days counter
                else:
                    consecutive_days = 0    # Reset the consecutive days counter if the gap is more than 1 day

                 # If an employee has worked for 7 consecutive days, print their name and position
                if consecutive_days == 6:   # Since we start counting from 0, 6 consecutive days means 7 in total
                    print(f"Employee Name: {employee}, Position: {row['Position ID']}, Consecutive Days: 7")
                    break
            prev_date = row['Time']    # Update the previous date with the current date

# now we can have our output by calling the function that we created above
find_employees_with_consecutive_days(df)

```

```python
# Now in second task we are printing the name and position of employees with short but reasonable time between shifts.
def find_employees_with_short_shifts(df):
    # Converting the 'Time' column to datetime
    df['Time'] = pd.to_datetime(df['Time'], errors='coerce')

    # Sort the DataFrame by 'Employee Name' and 'Time'
    df = df.sort_values(by=['Employee Name', 'Time'])

    # Define the acceptable time range between shifts
    min_hours_between_shifts = 1
    max_hours_between_shifts = 10

    # Iterate through the data to find employees with short shifts
    for employee, group in df.groupby('Employee Name'):
        prev_time = None
        for idx, row in group.iterrows():
            if prev_time is not None:
              # Calculate the time difference between the current timestamp and the previous timestamp in hours
                time_diff_hours = (row['Time'] - prev_time).total_seconds() / 3600

                if min_hours_between_shifts < time_diff_hours < max_hours_between_shifts:
                   # If the time difference falls within the specified range, print employee details
                    print(f"Employee Name: {employee}, Position: {row['Position ID']}, "
                          f"Time Between Shifts: {time_diff_hours:.2f} hours")
            prev_time = row['Time']  # Update the previous timestamp with the current timestamp


# now we can have our output by calling the function that we created above
find_employees_with_short_shifts(df)
```

```python
# Our final and third task is to print the name and position of employees who have worked for more than 14 hours in a single shift.
def find_employees_with_long_shifts(df):
    # Convert the 'Time Out' column to datetime
    df['Time Out'] = pd.to_datetime(df['Time Out'], errors='coerce')

    # Define the threshold for a long shift
    max_hours_per_shift = 14

    # Iterate through the data to find employees with long shifts
    for idx, row in df.iterrows():
      # Calculate the shift duration in hours for the current row
        shift_duration = (row['Time Out'] - row['Time']).total_seconds() / 3600

        if shift_duration > max_hours_per_shift:
             # If the shift duration is greater than the specified threshold, print employee details
            print(f"Employee Name: {row['Employee Name']}, Position: {row['Position ID']}, "
                  f"Shift Duration: {shift_duration:.2f} hours")


# now we can have our output by calling the function that we created above
find_employees_with_long_shifts(df)
```
